```python
import numpy as np
def plot_cat_rate(df,col):
    agg = df.groupby(col)['CASE_STATUS'].mean()
    agg_sum=df.groupby(col)['CASE_STATUS'].count()
    err=1.95*np.sqrt(agg*(1-agg)/agg_sum)
    agg.plot(kind='bar',yerr=err).set_ylabel('rejection rate')
    print(col)
    print(df_2017_status[col].value_counts())
```

```python
# Y/N to 1/0
df_00['AGENT_REPRESENTING_EMPLOYER'] = (df_00['AGENT_REPRESENTING_EMPLOYER'] == 'Y').astype(int)
```

```python
df_00['FULL_TIME_POSITION'] = (df_00['FULL_TIME_POSITION'] == 'Y').astype(int)
df_00['H1B_DEPENDENT'] = (df_00['H1B_DEPENDENT'] == 'Y').astype(int)
df_00['WILLFUL_VIOLATOR'] = (df_00['WILLFUL_VIOLATOR'] == 'Y').astype(int)
df_00['SUPPORT_H1B'] = (df_00['SUPPORT_H1B'] == 'Y').astype(int)
df_00['LABOR_CON_AGREE'] = (df_00['LABOR_CON_AGREE'] == 'Y').astype(int)
```

```python
df_00.columns
```

```python
df_00_inp=df_00[['AGENT_REPRESENTING_EMPLOYER', 'FULL_TIME_POSITION', 'H1B_DEPENDENT', 'WILLFUL_VIOLATOR', 'SUPPORT_H1B','LABOR_CON_AGREE']]
y=df_00.CASE_STATUS.values
```

```python
df_00_inp.info()
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_00_inp, y, test_size=0.30, random_state=67, stratify=y)
```

```python
#first model
rfc = RandomForestClassifier(n_estimators=500, n_jobs=-1)
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```

```python
#nice base line
pd.crosstab(y_test, rfc.predict(X_test))
```

```python
#random guess???
import sklearn.metrics as skm
import numpy as np
skm.roc_auc_score(y_test,rfc.predict_proba(X_test)[:, 1])
```
