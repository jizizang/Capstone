```python
X['H1B_RATE']=X.groupby('H1B_DEPENDENT')['CASE_STATUS'].transform('sum')/X.groupby('H1B_DEPENDENT')['H1B_DEPENDENT'].transform('count')
```

```python
X['WIL_RATE']=X.groupby('WILLFUL_VIOLATOR')['CASE_STATUS'].transform('sum')/X.groupby('WILLFUL_VIOLATOR')['WILLFUL_VIOLATOR'].transform('count')
```

```python
X['FULL_RATE']=X.groupby('FULL_TIME_POSITION')['CASE_STATUS'].transform('sum')/X.groupby('FULL_TIME_POSITION')['FULL_TIME_POSITION'].transform('count')
X['AGENT_RATE']=X.groupby('AGENT_REPRESENTING_EMPLOYER')['CASE_STATUS'].transform('sum')/X.groupby('AGENT_REPRESENTING_EMPLOYER')['AGENT_REPRESENTING_EMPLOYER'].transform('count')
X['SUPPORT_RATE']=X.groupby('SUPPORT_H1B')['CASE_STATUS'].transform('sum')/X.groupby('SUPPORT_H1B')['SUPPORT_H1B'].transform('count')
X['LABOR_RATE']=X.groupby('LABOR_CON_AGREE')['CASE_STATUS'].transform('sum')/X.groupby('LABOR_CON_AGREE')['LABOR_CON_AGREE'].transform('count')
X['LEVEL_RATE']=X.groupby('PW_WAGE_LEVEL')['CASE_STATUS'].transform('sum')/X.groupby('PW_WAGE_LEVEL')['PW_WAGE_LEVEL'].transform('count')
X['UNIT_RATE']=X.groupby('WAGE_UNIT_OF_PAY')['CASE_STATUS'].transform('sum')/X.groupby('WAGE_UNIT_OF_PAY')['WAGE_UNIT_OF_PAY'].transform('count')
```

```python
    group=dict(zip(X.H1B_DEPENDENT,X.H1B_RATE))
    X_test['H1B_RATE'] = X_test['H1B_DEPENDENT'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
    
    group=dict(zip(X.WILLFUL_VIOLATOR,X.WIL_RATE))
    X_test['WIL_RATE'] = X_test['WILLFUL_VIOLATOR'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
    
    group=dict(zip(X.FULL_TIME_POSITION,X.FULL_RATE))
    X_test['FULL_RATE'] = X_test['FULL_TIME_POSITION'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
    
    group=dict(zip(X.AGENT_REPRESENTING_EMPLOYER,X.AGENT_RATE))
    X_test['AGENT_RATE'] = X_test['AGENT_REPRESENTING_EMPLOYER'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
    
    group=dict(zip(X.SUPPORT_H1B,X.SUPPORT_RATE))
    X_test['SUPPORT_RATE'] = X_test['SUPPORT_H1B'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
    
    group=dict(zip(X.LABOR_CON_AGREE,X.LABOR_RATE))
    X_test['LABOR_RATE'] = X_test['LABOR_CON_AGREE'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
    
    group=dict(zip(X.PW_WAGE_LEVEL,X.LEVEL_RATE))
    X_test['LEVEL_RATE'] = X_test['PW_WAGE_LEVEL'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
    
    group=dict(zip(X.WAGE_UNIT_OF_PAY,X.UNIT_RATE))
    X_test['UNIT_RATE'] = X_test['WAGE_UNIT_OF_PAY'].map(group, na_action=None)
    X_test.fillna(y_train.mean(), inplace=True)
```
